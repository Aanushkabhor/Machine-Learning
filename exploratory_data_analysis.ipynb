```python
data = data.drop(['S.No.'], axis = 1)
data.info()
```

```python
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
data['Brand'] = data.Name.str.split().str.get(0)
```

```python
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
```

```python
data[['Name','Brand','Model']]
```

```python
data['Petrol'] = data.Fuel_Type.str.split().str.get(0)
```

```python
data['Disel'] = data.Fuel_Type.str.split().str.get(1) + data.Name.str.split().str.get(2)
```

```python
data[['Fuel_Type','Petrol','Disel']]
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
data['Price'] = pd.cut(data['Price'], 
bins=[0, 5000, 10000, 20000, 50000, 100000],
labels=['<5K', '5-10K', '10-20K', '20-50K', '50K+'])

# Then plot
data['Price'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Price Range Distribution')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))
plt.title('Bar Plot for Top Car Models')

# Bar plot for top 20 models (you can change the number)
top_models = data['Model'].value_counts().head(20).index
sns.countplot(x='Model', data=data[data['Model'].isin(top_models)],
              order=top_models, palette='Set2')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13,17))
sns.barplot(data=data.drop(['Name','Location'],axis=1))
plt.show()
```

```python
plt.scatter(data['Kilometers_Driven'],data['Car_Age'])
plt.xlabel("Kilometers")
plt.ylabel("Age")
plt.title("Age of Car From kilometers driven")
plt.show()
```

```python
counts=pd.crosstab(data['Fuel_Type'],data['Transmission'])
counts.plot(kind='bar')
plt.title("Fuel_Type VS Transmission")
plt.xlabel("Fuel_Type")
plt.ylabel("Transmission")
plt.xticks(rotation=0)
plt.legend(title="Transmission")
plt.tight_layout()
plt.show()
```
